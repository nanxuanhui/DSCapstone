```python
from transformers import pipeline
from langchain_community.utilities import SerpAPIWrapper, WikipediaAPIWrapper
from langchain_community.tools import WikipediaQueryRun, Tool
from langchain.agents import initialize_agent, AgentType
from langchain.llms import HuggingFacePipeline
import whisper  # Whisper for voice commands
import numpy as np
from IPython.display import Audio, display, Javascript  # For recording audio in Colab
import io
import soundfile as sf  # For saving audio
import base64
from google.colab import files  # For uploading audio files

# Set up Hugging Face summarization pipeline
summarizer = pipeline("summarization", model="philschmid/bart-large-cnn-samsum", device="cuda:0")

# Wrap the Hugging Face pipeline in a LangChain-compatible LLM
huggingface_llm = HuggingFacePipeline(pipeline=summarizer)

# Set up web search and Wikipedia tools
def search_with_timeout(query: str, timeout: int = 10) -> str:
    try:
        search_wrapper = SerpAPIWrapper(serpapi_api_key="0798336c05c0dfd4bec558798fbfc1544a3bec04cc9a037b3dedf1b16310ed55")
        result = search_wrapper.run(query)
        print("Search result:", result)  # Debugging line
        return result
    except Exception as e:
        print("Search error:", e)  # Debugging line
        return "Error: Search failed."

search_tool = Tool(
    name="Search",
    func=lambda query: search_with_timeout(f"{query} num_results=5"),
    description="Useful for searching the web for up-to-date information. Limits results to 5 items."
)

# Set up Wikipedia tool
wiki_api_wrapper = WikipediaAPIWrapper()
wiki_tool = WikipediaQueryRun(api_wrapper=wiki_api_wrapper)

# Define a function to summarize text using Hugging Face
def summarize_text(text: str, max_length: int = 150) -> str:
    try:
        # Join the list of search results into a single string
        if isinstance(text, list):
            text = " ".join(text)

        # Truncate the text to fit within the model's token limit (1024 tokens for BART)
        max_tokens = 1024
        truncated_text = " ".join(text.split()[:max_tokens])

        # Generate a summary with bullet points
        prompt = f"Summarize the following text into 5 bullet points:\n\n{truncated_text}"
        summary = summarizer(prompt, max_length=max_length, min_length=30, do_sample=False)
        return summary[0]['summary_text']
    except Exception as e:
        print("Summarization error:", e)
        return "Error: Summarization failed."

# Load Whisper model for voice commands
whisper_model = whisper.load_model("small")  # You can use "small", "medium", or "large" for better accuracy

# Function to record audio in Colab using JavaScript
def record_audio(duration=5, sample_rate=16000):
    # JavaScript code to record audio in the browser
    js_code = """
    const sleep = (ms) => new Promise(resolve => setTimeout(resolve, ms));
    const record = (duration, sampleRate) => {
        return new Promise((resolve) => {
            navigator.mediaDevices.getUserMedia({ audio: true }).then((stream) => {
                const mediaRecorder = new MediaRecorder(stream);
                const audioChunks = [];
                mediaRecorder.addEventListener("dataavailable", (event) => {
                    audioChunks.push(event.data);
                });
                mediaRecorder.start();
                sleep(duration * 1000).then(() => {
                    mediaRecorder.stop();
                    stream.getTracks().forEach((track) => track.stop());
                    const audioBlob = new Blob(audioChunks, { type: "audio/wav" });
                    const reader = new FileReader();
                    reader.readAsDataURL(audioBlob);
                    reader.onloadend = () => {
                        resolve(reader.result.split(",")[1]);
                    };
                });
            });
        });
    };
    record(%d, %d).then((data) => {
        google.colab.kernel.invokeFunction('notebook.saveAudio', [data], {});
    });
    """ % (duration, sample_rate)

    # Display the JavaScript code in Colab
    display(Javascript(js_code))

# Function to save the recorded audio
def save_audio(data):
    # Decode the base64 audio data
    audio_bytes = base64.b64decode(data)
    with open("recorded_audio.wav", "wb") as f:
        f.write(audio_bytes)
    print("Audio saved as recorded_audio.wav")

# Function to upload an audio file
def upload_audio():
    uploaded = files.upload()
    for filename in uploaded.keys():
        print(f"Uploaded {filename}")
        return filename
    return None

# Function to transcribe audio using Whisper
def transcribe_audio(audio_file):
    try:
        result = whisper_model.transcribe(audio_file)
        return result["text"]
    except Exception as e:
        print("Transcription error:", e)
        return "Error: Transcription failed."

# Register the save_audio function in Colab
from google.colab import output
output.register_callback('notebook.saveAudio', save_audio)

# Create an AI assistant agent
agent = initialize_agent(
    tools=[search_tool, wiki_tool],  # Search tool is first
    llm=huggingface_llm,  # Use the Hugging Face LLM
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    handle_parsing_errors=True,  # Handle parsing errors gracefully
    max_iterations=5  # Limit the number of iterations
)

# Define a refined prompt to guide the agent
refined_prompt = """
You are an AI assistant tasked with finding and summarizing information on any topic.
Use the Search tool to find relevant information and summarize the key insights in 5 bullet points.
"""

# Main loop for user interaction
while True:
    # Ask the user if they want to type or use voice
    input_method = input("Would you like to type or use voice? (type/voice/exit): ").strip().lower()

    # Exit condition
    if input_method in ["exit", "quit", "stop"]:
        print("Goodbye!")
        break

    if input_method == "type":
        user_input = input("What would you like to know about? ")
    elif input_method == "voice":
        # Ask the user if they want to record or upload an audio file
        voice_option = input("Would you like to record or upload an audio file? (record/upload): ").strip().lower()
        if voice_option == "record":
            print("Recording... Speak now!")
            record_audio(duration=5)  # Record for 5 seconds
            user_input = transcribe_audio("recorded_audio.wav")
        elif voice_option == "upload":
            print("Please upload an audio file.")
            audio_file = upload_audio()
            if audio_file:
                user_input = transcribe_audio(audio_file)
            else:
                print("No audio file uploaded. Please try again.")
                continue
        else:
            print("Invalid option. Please choose 'record' or 'upload'.")
            continue
        print(f"Transcribed input: {user_input}")
    else:
        print("Invalid input method. Please choose 'type', 'voice', or 'exit'.")
        continue

    # Step 1: Fetch information using SerpAPI
    search_query = f"{user_input} num_results=5"
    search_results = search_with_timeout(search_query)

    # Step 2: Summarize the search results using Hugging Face
    if not search_results.startswith("Error:"):
        summary = summarize_text(search_results)
        print("Final Summary:\n", summary)
    else:
        print("Failed to fetch search results. Please check your SerpAPI key and configuration.")
```

```python
from fastapi import FastAPI, HTTPException, File, UploadFile
from pydantic import BaseModel
from transformers import pipeline
from langchain_community.utilities import SerpAPIWrapper
from langchain_community.tools import Tool
import whisper
import io
import logging
import asyncio
import os

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Set up Hugging Face summarization pipeline
try:
    summarizer = pipeline("summarization", model="philschmid/bart-large-cnn-samsum", device="cuda:0")
    logger.info("Summarization pipeline loaded successfully.")
except Exception as e:
    logger.error(f"Failed to load summarization pipeline: {e}")
    summarizer = None

# Set up web search tool
def search_with_timeout(query: str, timeout: int = 10) -> str:
    try:
        search_wrapper = SerpAPIWrapper(serpapi_api_key="0798336c05c0dfd4bec558798fbfc1544a3bec04cc9a037b3dedf1b16310ed55")
        result = search_wrapper.run(query)

        # Check if the result is empty or invalid
        if not result or result.strip() == "":
            return "Error: No search results found."

        logger.info(f"Search results: {result}")
        return result
    except Exception as e:
        logger.error(f"Search failed: {e}")
        return f"Error: Search failed. {str(e)}"

search_tool = Tool(
    name="Search",
    func=lambda query: search_with_timeout(f"{query} num_results=5"),
    description="Useful for searching the web for up-to-date information. Limits results to 5 items."
)

# Load Whisper model for voice commands
try:
    whisper_model = whisper.load_model("small")
    logger.info("Whisper model loaded successfully.")
except Exception as e:
    logger.error(f"Failed to load Whisper model: {e}")
    whisper_model = None

# Validate input text length
def validate_input_text(text: str, min_length: int = 10, max_length: int = 1024) -> bool:
    word_count = len(text.split())
    if word_count < min_length:
        logger.warning(f"Input text is too short: {word_count} words.")
    elif word_count > max_length:
        logger.warning(f"Input text is too long: {word_count} words.")
    return min_length <= word_count <= max_length

# Define a function to summarize text using Hugging Face
def summarize_text(text: str, max_length: int = 150) -> str:
    try:
        if isinstance(text, list):
            text = " ".join(text)

        # Check if the text is too short
        if len(text.split()) < 10:
            return "Error: Input text is too short for summarization."

        # Truncate the text to fit within the model's token limit
        max_tokens = 1024
        truncated_text = " ".join(text.split()[:max_tokens])

        # Generate a summary with bullet points
        prompt = f"Summarize the following text into 5 bullet points:\n\n{truncated_text}"
        summary = summarizer(prompt, max_length=max_length, min_length=30, do_sample=False)
        return summary[0]['summary_text']
    except Exception as e:
        logger.error(f"Summarization failed: {e}")
        return f"Error: Summarization failed. {str(e)}"

# Function to transcribe audio using Whisper
def transcribe_audio(audio_file):
    try:
        result = whisper_model.transcribe(audio_file)
        return result["text"]
    except Exception as e:
        logger.error(f"Transcription failed: {e}")
        return f"Error: Transcription failed. {str(e)}"

# Initialize FastAPI app
app = FastAPI()

# Define a Pydantic model for the request body
class QueryRequest(BaseModel):
    query: str

# Health check endpoint
@app.get("/health")
async def health_check():
    status = {
        "whisper_model_loaded": whisper_model is not None,
        "summarizer_loaded": summarizer is not None,
        "serpapi_available": True,  # Add a test query to verify SerpAPI
    }
    return status

# API endpoint for searching and summarizing
@app.post("/search_and_summarize")
async def search_and_summarize(request: QueryRequest):
    query = request.query
    if not query:
        raise HTTPException(status_code=400, detail="Please provide a valid query.")

    if not summarizer:
        raise HTTPException(status_code=500, detail="Summarization service is currently unavailable.")

    # Perform search
    search_results = await asyncio.to_thread(search_with_timeout, query)
    if search_results.startswith("Error:"):
        raise HTTPException(status_code=400, detail=search_results)

    # Validate input text length
    if not validate_input_text(search_results):
        raise HTTPException(status_code=400, detail="Search results are too short or too long for summarization.")

    # Summarize the search results
    summary = await asyncio.to_thread(summarize_text, search_results)
    if summary.startswith("Error:"):
        raise HTTPException(status_code=500, detail=summary)

    return {"summary": summary}

# API endpoint for transcribing audio
@app.post("/transcribe_audio")
async def transcribe_audio_endpoint(file: UploadFile = File(...)):
    if not whisper_model:
        raise HTTPException(status_code=500, detail="Transcription service is currently unavailable.")

    try:
        # Log that a file has been received
        logger.info(f"Received file: {file.filename}, Content-Type: {file.content_type}")

        # Check if the file is valid
        if not file.filename or not file.content_type.startswith("audio/"):
            raise HTTPException(status_code=400, detail="Invalid file format. Please upload an audio file.")

        # Save the uploaded file temporarily
        file_path = f"temp_{file.filename}"
        with open(file_path, "wb") as buffer:
            buffer.write(await file.read())

        # Log file size for debugging
        file_size = os.path.getsize(file_path)
        logger.info(f"File size: {file_size} bytes")

        # Log that the file is being processed
        logger.info("Processing audio file...")

        # Transcribe the audio file
        transcription = await asyncio.to_thread(transcribe_audio, file_path)

        # Log that the file has been processed
        logger.info("Audio file processed successfully.")

        # Clean up the temporary file
        os.remove(file_path)

        if transcription.startswith("Error:"):
            raise HTTPException(status_code=400, detail=transcription)

        return {"transcription": transcription}
    except Exception as e:
        logger.error(f"Error processing audio: {e}")
        raise HTTPException(status_code=500, detail=f"Error processing audio: {str(e)}")

# Run FastAPI app
if __name__ == "__main__":
    # Start Ngrok tunnel
    ngrok_tunnel = ngrok.connect(8000)
    print("Public URL:", ngrok_tunnel.public_url)

# Run FastAPI app
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
    import uvicorn
```
